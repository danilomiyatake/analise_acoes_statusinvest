```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
codigo = []
nome_empresa = []
cnpj = []
segmento = []
setor = []

headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36 Edg/108.0.1462.46',
}
    

def get_request(url_base, headers):
    return requests.get(url_base, headers=headers)


for i in range(1,12):
    response = get_request(f'https://maisretorno.com/lista-acoes/page/{i}', headers)
    
    response_text = response.text
    
    soup = BeautifulSoup(response_text, 'lxml')
    
    results = soup.find_all('li', class_='MuiListItem-root MuiListItem-gutters MuiListItem-padding MuiListItem-divider css-vw54ay')
    
    for result in results:
        divisoes = result.div.find_all('div')
        codigo.append(divisoes[0].a.text)
        nome_empresa.append(divisoes[0].p.text)
        cnpj.append(divisoes[1].h6.text)
        segmento.append(divisoes[2].h6.text)
        setor.append(divisoes[3].h6.text)
```

```python
info = {
    'codigo' : codigo,
    'nome da empresa' : nome_empresa,
    'cnpj' : cnpj,
    'segmento' : segmento,
    'setor' : setor
}
```

```python
df = pd.DataFrame(info)
df.isnull().sum()
```

```python
df.drop(index = df[df['cnpj'] == '-'].index, axis=0, inplace=True)
```

```python
df.to_csv('lista_empresa.csv', sep = ';', index=False, header=True )
```
